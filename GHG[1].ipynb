```python
# create an empty dataframe to store the forecast results
forecast_all = pd.DataFrame()

# set the number of iterations
num_iterations = 20

# iterate over the code for the specified number of times
for i in range(num_iterations):
    # create a copy of the original dataframe to concatenate with the forecast dataframe
    ghg_2022_final_copy = ghg_2022_final.copy()
    
    # create an empty dataframe to store the forecast results for this iteration
    forecast_df = pd.DataFrame(columns=ghg_2022_final.columns)
    
    # iterate over the columns in the original dataframe
    for column in ghg_2022_final_copy.columns:
        # Extract the column data as y
        y = ghg_2022_final_copy[column]

        # Fit the SARIMAX model with the selected column as the endogenous variable
        model = SARIMAX(y, order=(1, 1, 1))
        results = model.fit()

        # Generate forecast for the next year
        forecast = results.forecast(steps=1)

        # Add the forecast value to the forecast_df dataframe
        forecast_df[column] = forecast
    
    # concatenate the forecast dataframe with the original dataframe
    ghg_2022_final_copy = pd.concat([ghg_2022_final_copy, forecast_df])
    
    # concatenate the forecast dataframe with the previous forecast dataframes
    forecast_all = pd.concat([forecast_all, forecast_df])
    
    # use the concatenated dataframe for the next iteration
    ghg_2022_final = ghg_2022_final_copy

# view the final concatenated forecast dataframe
print(forecast_all)
```
